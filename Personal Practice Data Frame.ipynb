```python
df_siswa['Tinggi Badan (cm)'] = df_siswa['Tinggi Badan (cm)'] / 100
df_siswa.head()
```

```python
df_siswa.rename(columns = {'Tinggi Badan (cm)' : 'Tinggi Badan (m)'}, inplace = True)
df_siswa.head()
```

```python
df_siswa[" BMI "] = df_siswa ['Berat badan (kg)'] / df_siswa [ "Tinggi Badan (m)"] ** 2
df_siswa.head()
```

```python
def status_BMI(a):
    if (a < 18.50):
        return 'Underweight'
    elif (a >= 18.50 and a <= 24.90):
        return 'Normal'
    elif (a >= 25 and a <= 29.90):
        return 'Overweight'
    elif (a >= 30 and a <= 34.90):
        return 'Obesity'
    elif (a >= 34.90):
        return 'Severly Obesity'
    else:
        return 'None'

df_siswa["Status BMI"] = df_siswa[" BMI "].apply(status_BMI)
df_siswa.head()
```

```python
df_Tinggi = df_siswa.sort_values(
    by = "Tinggi Badan (m)" ,
    ascending=False)

df_Tinggi.head()
```
